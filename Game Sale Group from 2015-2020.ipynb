```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
from matplotlib import cm
from collections import OrderedDict
import plotly.graph_objects as go
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
df = pd.read_csv("vgsales_clean_v2.csv")
df
```

```python
df_15_20 = df[df['Year']>=2015]
df_15_20 = df_15_20.sort_values(by="Global_Sales", ascending=False)
df_15_20
```

```python
sales_list = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']


for i in sales_list:
  df_15_20_tmp = df_15_20.sort_values(by=i, ascending=False)
  plt.figure(figsize=(5, 7))
  plt.bar(df_15_20_tmp["Name"][:7], df_15_20_tmp[i][:7])
  plt.xticks(df_15_20_tmp["Name"][:7], wrap=True, rotation=90, fontsize=15)
  #plt.xlabel("Games")
  plt.ylabel(i + "\n\n (millions)", fontsize=25)
  #plt.title(i)
  plt.show()
```
